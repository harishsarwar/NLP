```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('spam.csv')
df.head()
```

```python
df.shape
```

```python
df.Category.value_counts()
```

```python
# creating new column for spam
df['spam'] = df['Category'].apply(lambda x : 1 if x =='spam' else 0)
```

```python
df.head()
```

```python
# same think with function
def spam(x):
    if x == 'spam':
        return 1
    return 0
```

```python
df['spam'] = df.Category.apply(spam)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.Message
y = df.spam
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

 # CountVectorizer helps to make BAG TO WORD model for Numeric repersentation.
# Documentation of CountVectorizer = https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html
```

```python
type(X_train.values)

# value : going to convert in to the numpy array
```

```python
v = CountVectorizer()

X_train_CountVectorizer = v.fit_transform(X_train.values)
# fit_transform is method which makes the BAG TO WORD model for X_train
X_train_CountVectorizer
```

```python
# values show in to spare matrix
# converting sparce matrix into numpy array.

X_trian_np = X_train_CountVectorizer.toarray()
```

# all set for machine learning model

```python
from sklearn.naive_bayes import MultinomialNB
model =  MultinomialNB()
model.fit(X_trian_np,y_train)
```

```python
X_trian_np.shape
```

```python
# Countvectorizer for X_test
X_test_cv = v.transform(X_test)
X_test_cv = X_test_cv.toarray()
X_test_cv[:4]
```

```python
X_test_cv.shape
```

```python
X_trian_np.shape
```

```python
prediction = model.predict(X_test_cv)
prediction
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,prediction))
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]

email_count= v.transform(emails)
model.predict(email_count)
```

# Pipe Line

```python
# same code with piple line for reduce the time and number of lines of code
```

```python
from sklearn.pipeline import Pipeline

clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('naive_base', MultinomialNB())
])
```

```python
clf.fit(X_train,y_train)
```

```python
prediction= clf.predict(X_test)
```

```python
print(classification_report(y_test,prediction))
```
